```python

spending = []
spend_summ = 0
will_remain = []

for i in range(10000):
  spending.append(list(input('Введите дату и время траты в формате YYYY-MM-DD hh:mm:ss S:    ').split()))
  if spending[i] == []:
    break

  spend_summ = spend_summ + int(spending[i][2])
  Time_past = (int(spending[i][0][8:10]) -1)*24*60*60 + int(spending[i][1][0:2])*60*60 + int(spending[i][1][3:5])*60 + int(spending[i][1][6:8])
  Time = 31*24*60*60 - Time_past
  if Time_past==0:
    will_remain = 0
  else:
    will_remain = round(10000 - spend_summ - spend_summ/Time_past*Time, 2)
  print("%.2f" % will_remain)
  
```

```python
spending = []
spend_summ = 0
will_remain = []
spending = list(iter(input('Введите дату и время траты в формате YYYY-MM-DD hh:mm:ss S:    ').split()))
for i in range(round(len(spending)/3)):

  spend_summ = spend_summ + int(spending[i*3+2])
  Time_past = (int(spending[i*3][8:10]) -1)*24*60*60 + int(spending[i*3+1][0:2])*60*60 + int(spending[i*3+1][3:5])*60 + int(spending[i*3+1][6:8])
  Time = 31*24*60*60 - Time_past
  if Time_past==0:
    will_remain = round(10000 - spend_summ - spend_summ*Time, 2)
  else:
    will_remain = round(10000 - spend_summ - spend_summ/Time_past*Time, 2)
  print("%.2f" % will_remain)
```
