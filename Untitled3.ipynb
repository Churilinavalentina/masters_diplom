Входной датасет

```python
from math import pi, sin, cos
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

tt = []
uu = []
hh = []
x = []
y = []
z = []
color = []
for i in range(31):
  tt.append((-1 + i*0.05)*pi)
  uu.append((0.5-i*0.05)*pi)
  if i < 20:
    hh.append(i*0.05)

for column in range(len(hh)):
  for line in range(2*len(tt)):
    if line < len(tt):
      x.append(cos(tt[line]))
      z.append(sin(tt[line]))
      color.append(tt[line])
    else:
      x.append(-cos(uu[line-len(tt)]))
      z.append(2-sin(uu[line-len(tt)]))
      color.append(uu[line-len(tt)])
    y.append(hh[column])
    
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=color)
print(len(x))
```

```python
# Второй датасет

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = [0.5, 0.4, 0.2, 0.8, 0.1, 0.9, 0.7, 0.3, 0.2, 0.1, 4.1, 4.5, 4.8, 4.3, 4.6, 4.5, 4.9, 4.1, 4.6, 4.2]
y = [0.1, 0.2, 0.9, 0.7, 0.3, 0.6, 0.4, 0.2, 0.8, 0.4, 4.9, 4.5, 4.6, 4.2, 4.6, 4.9, 4.8, 4.2, 4.3, 4.7]
z = [0.2, 0.8, 0.3, 0.4, 0.5, 0.7, 0.1, 0.6, 0.5, 0.7, 4.1, 4.6, 4.8, 4.3, 4.9, 4.6, 4.9, 4.1, 4.2, 4.9]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
```

```python
K = 8

# Поиск ближайших соседей

neighbours_coord_x = []
neighbours_coord_y = []
neighbours_coord_z = []
x_copy = x.copy()
y_copy = y.copy()
z_copy = z.copy()
N = 8
x_copy.pop(N)
y_copy.pop(N)
z_copy.pop(N)
x_N = x[N]
y_N = y[N]
z_N = z[N]


def neighbours(N, x_N, y_N, z_N, x_copy, y_copy, z_copy, K, neighbours_coord_x, neighbours_coord_y, neighbours_coord_z):
  min_distance = 1000

  for i in range(len(x_copy)):
    distance = ((x_copy[i]-x_N)**2 + (y_copy[i]-y_N)**2 + (z_copy[i]-z_N)**2)**1/2
    if distance < min_distance:
      min_distance = distance
      neighbours_number = i

  neighbours_coord_x.append(x_copy[neighbours_number])
  neighbours_coord_y.append(y_copy[neighbours_number])
  neighbours_coord_z.append(z_copy[neighbours_number])
  x_copy.pop(neighbours_number)
  y_copy.pop(neighbours_number)
  z_copy.pop(neighbours_number)

  if len(neighbours_coord_z) < K:
    neighbours(N, x_N, y_N, z_N, x_copy, y_copy, z_copy, K, neighbours_coord_x, neighbours_coord_y, neighbours_coord_z)
  return neighbours_coord_x, neighbours_coord_y, neighbours_coord_z


neighbours_coord_x, neighbours_coord_y, neighbours_coord_z = neighbours(N, x_N, y_N, z_N, x_copy, y_copy, z_copy, K, neighbours_coord_x, neighbours_coord_y, neighbours_coord_z)

print(neighbours_coord_x, neighbours_coord_y, neighbours_coord_z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(neighbours_coord_x, neighbours_coord_y, neighbours_coord_z)
print(len(x))
```

находим W, решая систему линейных уравнений

```python
import numpy as np

def search_W(x_N, y_N, z_N, neighbours_coord_x, neighbours_coord_y, neighbours_coord_z):
  W = []
  for i in range(len(neighbours_coord_x)):
    W.append(0)

  A = np.vstack([neighbours_coord_x, neighbours_coord_y, neighbours_coord_z])
  y = np.array([0, 0, 0])
  W  = np.linalg.lstsq(A, y, rcond=None)[0]
  
  return(W)

search_W(x_N, y_N, z_N, neighbours_coord_x, neighbours_coord_y, neighbours_coord_z)
```
